```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def plot_similarity(labels, features, rotation=90):
    # Compute inner product similarity matrix
    corr = np.inner(features, features)

    # Set style
    sns.set(font_scale=1.2)

    # Plot heatmap
    g = sns.heatmap(
        corr,
        xticklabels=labels,
        yticklabels=labels,
        vmin=0,
        vmax=1,
        cmap="YlOrRd"
    )

    # Rotate labels if needed
    plt.xticks(rotation=rotation)
    plt.yticks(rotation=0)
    plt.show()
```

```python
# Install required libraries
! pip install tensorflow tensorflow-hub seaborn

# Import libraries
import tensorflow as tf
import tensorflow_hub as hub
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Universal Sentence Encoder
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model = hub.load(module_url)

print("Model Loaded Successfully!")

# Function to generate embeddings
def embed(input_texts):
    return model(input_texts)

# Load CSV file
df = pd.read_csv("D:\\Agentic_AI_Internship\\arxiv.csv")

# Check first few rows
print(df.head())

# Rename columns if needed
df.columns = ["Title", "Abstract"]

# Select first 20 titles for similarity analysis
messages = df["Title"].head(20).tolist()

# Generate embeddings
message_embeddings = embed(messages).numpy()

# Print embedding information
for i, embedding in enumerate(message_embeddings):
    print("Title:", messages[i])
    print("Embedding Size:", len(embedding))
    print("First 3 Values:", embedding[:3])
    print("-" * 50)

# Similarity Heatmap Function
def plot_similarity(labels, features, rotation=90):

    # Similarity Matrix
    corr = np.inner(features, features)

    plt.figure(figsize=(12, 10))

    sns.heatmap(
        corr,
        xticklabels=labels,
        yticklabels=labels,
        cmap="YlOrRd",
        vmin=0,
        vmax=1
    )

    plt.xticks(rotation=rotation)
    plt.yticks(rotation=0)

    plt.title("Paper Title Similarity Heatmap")
    plt.show()

# Plot Heatmap
plot_similarity(messages, message_embeddings)
```

```python
import spacy
import tabulate
from IPython.display import display, HTML

# Load English model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "Artificial Intelligence is transforming the world."

# Process text
doc_en = nlp(text)

# Extract features
features = [
    [t.text, t.lemma_, t.pos_, t.tag_]
    for t in doc_en
]

# Display table
display(
    HTML(
        tabulate.tabulate(
            features,
            headers=[
                "Word",
                "Lemma",
                "POS",
                "Tag"
            ],
            tablefmt='html'
        )
    )
)
```
