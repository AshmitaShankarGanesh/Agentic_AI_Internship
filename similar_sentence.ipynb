```python
import csv
import random

f = "D:\\Agentic_AI_Internship\\data\\movie-dialog-corpus\\movie_lines.tsv"

texts = []
with open(f) as i:
    csv_reader = csv.reader(i, delimiter="\t")
    for line in csv_reader:
        texts.append(line[-1])

random.seed(1)
random.shuffle(texts)
texts = texts[:10000]
texts[:10]        
```

```python
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np

embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/2")
```

```python
import tensorflow as tf
import numpy as np

def embed_texts(texts):

    embeddings = embed.signatures['default'](
        tf.constant(texts)
    )

    print(embeddings.keys())

    return embeddings

embeddings = embed_texts(texts)
```

```python
import tensorflow as tf
import numpy as np

def embed_texts(texts):

    embeddings = embed.signatures['default'](
        tf.constant(texts)
    )['default']

    return embeddings.numpy().tolist()

embeddings = embed_texts(texts)

print(len(embeddings))
print(len(embeddings[0]))
```
