Importing the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

Defining the Features(X) and Target(Y) for my data

```python
X = df.drop(['Salary', 'Unnamed: 0'], axis=1)
Y = df["Salary"]
```

Splitting the dataset into testing and training part

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

Model Selection

```python
model=LinearRegression()
```

Train Linear Regression Model

```python
model1 = model.fit(X_train,Y_train)
```

```python
model1
```

```python
predicted_salary = model.predict(X_test)
```

```python
predicted_salary
```

Model Evaluation

```python
errors = mean_squared_error(Y_test, predicted_salary)
print("Mean Squared Error:", errors)
r2 = r2_score(Y_test, predicted_salary)
print("R-squared:", r2)
mae = mean_absolute_error(Y_test, predicted_salary)
print("Mean Absolute Error:", mae)  
```
